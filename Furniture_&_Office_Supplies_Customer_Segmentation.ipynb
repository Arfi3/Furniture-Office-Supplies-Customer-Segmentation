# SEGMENTASI CUSTOMER



*  Segmentasi Customer Berbasis Statistik.

    * Order Segment: Jumlah transaksi yang dilakukan oleh customer (Total SalesOrderNumber).
    * Quantity Segment: Total jumlah barang yang dibeli (Order Quantity).
    * Sales Segment: Total nilai pembelian customer, dihitung dengan rumus:**(Order Quantity × Unit Price) − Discount.**

* Segmentasi Customer dengan Metode RFM.



```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/Arfi3/Office_Sales_Analysis/main/Sales.xlsx"
df = pd.read_excel(url)

df.head()
```

```python
url2="https://raw.githubusercontent.com/Arfi3/Office_Sales_Analysis/main/Customer.xlsx"
df_2 = pd.read_excel(url2)

df_2.head()
```

```python
df.nunique()
```

```python
df['Total_Purchase'] = df['UnitPrice'] *df['OrderQuantity']* (1 - df['Discount %'])
```

```python
df.head()
```

```python
df_3 = pd.merge(df, df_2, on='CustomerKey', how='left')
```

```python
df_3.head()
```

```python
df_3.info()
```

```python
# =========================
# AGREGASI PER CUSTOMER
# =========================
customer_agg = df_3.groupby('Customer').agg({
    'SalesOrderNumber': 'nunique',
    'OrderQuantity': 'sum',
    'Total_Purchase': 'sum'
}).reset_index()

customer_agg.columns = ['Customer', 'Total_Order', 'Total_Quantity', 'Total_Sales']
```

```python
customer_agg.head()
```

```python
# =========================
# SEGMENTASI STATISTIK
# =========================
def label_tier(series, labels):
    return pd.qcut(series, q=3, labels=labels, duplicates='drop')

# Total Order
customer_agg['Order_Segment'] = label_tier(
    customer_agg['Total_Order'],
    ['Silver', 'Gold', 'Platinum']
)

# Quantity
customer_agg['Quantity_Segment'] = label_tier(
    customer_agg['Total_Quantity'],
    ['Regular', 'Champion', 'King']
)

# Sales
customer_agg['Sales_Segment'] = label_tier(
    customer_agg['Total_Sales'],
    ['Prime', 'Elite', 'Diamond']
)

# =========================
# HASIL
# =========================
customer_agg.head()
```

```python
# =========================
# SEGMENTASI DENGAN RFM
# =========================
df_3['OrderDate'] = pd.to_datetime(df_3['OrderDate'])

# ambil tanggal terakhir di dataset
latest_date = df_3['OrderDate'].max()

# hitung recency (dalam hari)
rfm = df_3.groupby('Customer').agg({
    'OrderDate': lambda x: (latest_date - x.max()).days,
    'SalesOrderNumber': 'nunique',
    'Total_Purchase': 'sum'
}).reset_index()

rfm.columns = ['Customer', 'Recency', 'Frequency', 'Monetary']
```

```python
rfm.head()
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], q=3, labels=[3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'], q=3, labels=[1,2,3])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], q=3, labels=[1,2,3])
```

```python
rfm['RFM_Total'] = rfm[['R_Score','F_Score','M_Score']].astype(int).sum(axis=1)
```

```python
rfm['Segment'] = pd.qcut(
    rfm['RFM_Total'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
rfm.head()
```

```python
segmentasi_customer= pd.merge(customer_agg, rfm, on='Customer', how='inner')
```

```python
segmentasi_customer.head()
```

```python
segmentasi_customer.to_csv('customer_segmentation.csv', index=False)
```
